```python
hours = []
for dt in datetimes:
    time = dt.split(' ')[1]
    hour = int(time.split(':')[0])
    hours.append(hour)
```

```python
dates = []
for dt in datetimes:
    date = dt.split(' ')[0]
    dates.append(date)
```

```python
dates[:5]
```

```python
collections.Counter(dates).most_common(20)
```

```python
by_date = collections.Counter(dates).most_common(30)
```

```python
counts = []
for r in by_date:
    counts.append(r[1])
```
